```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Definindo o Ã¢ngulo ðœƒ de forma randÃ´mica entre 0 e ðœ‹
theta = random.uniform(0, np.pi)

# Calculando seno, cosseno e tangente de ðœƒ
sen_theta = np.sin(theta)
cos_theta = np.cos(theta)
tan_theta = np.tan(theta)

# Limitando a tangente para evitar valores muito grandes
tan_length = 2 if np.abs(tan_theta) > 2 else np.abs(tan_theta)

fig, ax = plt.subplots()

# Definindo o cÃ­rculo trigonomÃ©trico
circle = plt.Circle((0, 0), 1, color='blue', fill=False)

# Adicionando o cÃ­rculo ao grÃ¡fico
ax.add_artist(circle)

# Desenhando os eixos cartesianos
plt.axhline(0, color='black')
plt.axvline(0, color='black')

# Desenhando as linhas para seno, cosseno e tangente
plt.plot([0, cos_theta], [0, 0], label='Cosseno', color='green')
plt.plot([cos_theta, cos_theta], [0, sen_theta], label='Seno', color='red')
if tan_theta > 0:
    plt.plot([cos_theta, 1], [sen_theta, tan_length], label='Tangente', color='orange', linestyle='--')
else:
    plt.plot([cos_theta, 1], [sen_theta, -tan_length], label='Tangente', color='orange', linestyle='--')

# Ajustando limites do grÃ¡fico
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.text(1.05, -0.1, f'Î¸ = {np.degrees(theta):.2f}Â°', fontsize=12, transform=ax.transAxes)
plt.tight_layout()
plt.show()
```
